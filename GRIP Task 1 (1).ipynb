#### Let us begin by importing the necessary libraries:

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn import metrics
```

```python
from sklearn.model_selection import train_test_split
```

#### Let's read the dataset into a dataframe

```python
df = pd.read_csv('http://bit.ly/w-data')
df.head()
```

#### Dataset Exploration: Finding the dimensions of the dataset

```python
df.shape
```

#### Dataset Exploration: Obtaining the description of the dataset 

```python
df.describe()
```

#### Dataset Exploration: Checking if there are any null values in the given dataset

```python
df.isnull().sum()
```

##### Therefore we can clearly observe that there are no null values of both the attributes; Hours and Scores

#### Visualizing the data 

```python
plt.figure()
plt.plot(df['Hours'].values, df['Scores'].values, 'o', color='green')
plt.grid(True)
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Study Hours vs Scores')
plt.show()
```

##### A clear cut analysis from the above is that their is almost a linear relationship between the scores of the student and the study hours. As the number of hours increase, so does the score. 

#### Building a linear reggression model for the same

```python
X = df.iloc[:, :1].values  
y = df.iloc[:,-1:].values 
```

##### Here we segregated the data into dependent and independent variables respectively 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
model = LinearRegression()
```

#### Training the model

```python
model.fit(X_train, y_train)  
line = model.coef_*X+model.intercept_
```

#### Visualizing the linear regression model

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, line, color = 'green')
plt.grid(True)
plt.xlabel('Hours')
plt.ylabel('Score')
plt.title('Study Hours vs Scores')
plt.show()
```

#### Prediction model based on the above

```python
regr_predict = model.predict(X_test)
```

##### We will predict the scores for someone who studies for a time of 9.25 hours

```python
hours =  [[9.25]]
result = model.predict(hours)
print("Number of Hours studied by the student = {}".format(hours))
print("Score Prediction for the student = {}".format(result[0]))
```

##### As we can clearly see the score predicted is 95.23 (approx.)

#### Evaluation of the model: Error Basis

```python
print('Error in the model is:', metrics.mean_absolute_error(y_test, regr_predict))
```

##### Therefore the model thus built gives an absolute mean error of 7.883 (Approx.)
